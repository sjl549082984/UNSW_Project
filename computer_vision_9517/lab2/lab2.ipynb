```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

imgA = cv2.imread('wagon.jpg', 0)
imgB = cv2.imread('wheel.png', 0)
imgC = cv2.imread('toy.png', 0)




def cv_show(img, name):
    cv2.imshow(name, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


sift = cv2.xfeatures2d.SIFT_create(500)
kpsA, dpA = sift.detectAndCompute(imgA, None)
kpsB, dpB = sift.detectAndCompute(imgB, None)
kpsC, dpC = sift.detectAndCompute(imgC, None)


bf = cv2.BFMatcher()

matches = bf.match(dpA, dpB)
matches2 = bf.match(dpA, dpC)
print(len(matches))
print(len(matches2))

matches = sorted(matches, key=lambda x: x.distance)

compare1 = len(matches)/len(kpsB)
compare2 = len(matches2)/len(kpsC)
```

```python
# question a
imgp1 = cv2.drawKeypoints(imgA, kpsA, None, (255, 0, 0), 4)
plt.imshow(imgp1),plt.show()
```

```python
imgp2 = cv2.drawKeypoints(imgB, kpsB, None, (255, 0, 0), 4)
plt.imshow(imgp2),plt.show()
```

```python
imgp3 = cv2.drawKeypoints(imgC, kpsC, None, (255, 0, 0), 4)
plt.imshow(imgp3),plt.show()
```

```python
if(compare2 > compare1):
    ret = cv2.drawMatches(imgA, kpsA, imgB, kpsB, matches[:100], None, flags=2)
    cv2.imshow('ret', ret)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    ret = cv2.drawMatches(imgA, kpsA, imgC, kpsC, matches[:100], None, flags=2)
    cv2.imshow('ret', ret)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
